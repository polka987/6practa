Цель: Непосредственно изучить библиотеку matplotlib в языке программирования Python, а именно её раздел .pyplot. Он позволит нам визуально оторазить некоторые данные, например, из датасета. Также он позволит нам провести не только более глубокий анализ тех же данных, но и выявить некоторые их закономерности роста и ументшения. Помимо этого matplotlib также даст возможность вывить возможные звисимости данных другн от друга. Наконец, необходимо будет составить около 13-15 диаграмм разных видов, чтобы не только в теории, но и на практике изучить и закрепить изученный материал.

```python
import pandas as pd
import numpy as mp
import matplotlib.pyplot as plot
animes = pd.read_csv("anime.csv")
plot.figure(figsize = (10, 8))
plot.bar(animes['#'].head(55), animes['rating'].head(55), label = 'Количество рейтинг', color = 'green')
plot.title("количество персонажей в аниме")
plot.xlabel('ID аниме')
plot.ylabel('рейтинг')
```

н заданной диаграмме можно лицезреть рейтинг в анимешках 1 по 55. Это непосредственно поможет нм сделать некоторые выводы: например, что в первых 55 аниме рейтинг довольно высок и блиизтся к 10. Так в принципе должно быть у каждого аниме, если это не хентай. В потверждение слов создадим ещё одну диграмму рейтингов (с перонажами для разнообразия), но при этом возмём не первые, а последни, там как раз есть хентыч.

```python
import pandas as pd
import numpy as mp
import matplotlib.pyplot as plot
animes = pd.read_csv("anime.csv")
plot.figure(figsize = (10, 8))
plot.bar(animes['#'].tail(55), animes['rating'].tail(55), label = 'Количество рейтинг', color = 'green')
plot.title("количество персонажей в аниме")
plot.xlabel('ID аниме')
plot.ylabel('рейтинг')
```

как видно, рейтинг хентая, хоть и преувеличен (полюбому извращги накрутили со своих твинков), но всё же намного ниже, чем у нормального аниме, за режким исключением (кто знает, может там и не хентай вовсе)

```python
import pandas as pd
import numpy as mp
import matplotlib.pyplot as plot
animes = pd.read_csv("anime.csv")
plot.figure(figsize = (10, 10))
plot.bar(animes['#'].head(100), animes['members'].head(100), label = 'Количество персонажей', color = 'green')
plot.title("количество персонажей в аниме")
plot.xlabel('ID аниме')
plot.ylabel('количество персонажей')
```

как виндо, у первых 100 анимешек сильно колеблется количесвто персонажей, учавствующих в действиях. например, есть как минимум 2 аниме с практически запредельным количеством персонажей, в то время как в остальный это колиество придерживается середины, иногда падая до пола, как и мои надежды на продолжение любимых тайтлов.

```python
Fantasy_counts = animes[animes['type'] == 'TV'].head(55).value_counts()
plot.pie(Fantasy_counts, autopct='%1.1f%%', colors=['pink', 'blue'])
plot.title('соотношение фентези и других тайтлов')
plot.show()
```

В даанном же случае соритруются те тайтлы, которые вызодят в формате TV и нет. И это прекрасно.

```python
Fantasy_counts = animes[animes['genre'] == 'Fantasy'].head(50).value_counts()
plot.pie(Fantasy_counts, autopct='%1.1f%%', colors=['yellow', 'black'])
plot.title('соотношение фентези и других тайтлов')
plot.show()
```

Покажет вам отношение фентезийных, а значит лучших в априоре, аниме по отношению к другим тоже аниме. Розовый - это фентези

```python
plot.hist(animes[animes['genre'] == 'Hentai']['rating'].tail(55), bins = 40, color = 'yellow', label='Как рейтинг в хентыче распределяется', edgecolor = 'green')
plot.legend()
plot.show()
```

Диаграмма доказывает, что мои слова были верны, поскольку зачастую данная ересь имеет рейтинг = 5 иниже. Лишь редкие иссключения имеют рейтинг выше 6 и 8.

```python
plot.hist(animes[animes['type'] == 'OVA']['members'].tail(45), bins=40, color = 'purple', label = 'то, как разделены количество участников в OVA')
plot.xlabel('количества персонажей')
plot.ylabel('количества тайтлов')
plot.legend()
plot.show()
```

Таким образом мы  вычислили, что в OVA анимешка (как минимум в первых 45) наиболее часто встречается количество персонажей, не превышающее 225 - 250. Более высокие значения данного параметра встречаются реже, если вообще встречаются

```python
plot.scatter(animes[animes['genre'] == 'Hentai']['members'].head(60), animes[animes['genre'] == 'Hentai']['rating'].head(60), color = 'lightgreen')
plot.xlabel('количество персонажей')
plot.ylabel('Рейтинг')
plot.title('Зависимость рейтинга от роста кол-ва персонажей ')
```

Данный график показывает сущность грешных деяний и мыслей человека: чем больше персонажей учавствует в нарисованном мультике эротического характера, чем калоритнее состав, тем больше это нравиться людям. восточные дядьки явно знают больше нас.

```python
plot.scatter(animes[animes['type'] == 'TV']['episodes'].head(60), animes[animes['genre'] == 'Hentai']['rating'].head(60), color = 'lightgreen')
plot.xlabel('количество эпизодов')
plot.ylabel('Рейтинг')
plot.title('Зависимость рейтинга от роста кол-ва эпизодов ')
```

Это же в свою очередь показывает, что чем больше надо посмотреть, чтобы понять глубинный смысл тайтла, тем меньше заинтересованности он вызывает. Это показывает истинную натуру человека ленивого и трусливого: он боится всего, требющего больше, чем беглый взгляд, например, подробного изучени
я.

```python

plot.plot(animes['#'].tail(55), animes['episodes'].tail(55), label = 'кол-во эпизодов')
plot.title('Статистика эпзодов')
plot.ylabel('Кол - во эпизодов')
plot.xlabel('ID аниме')
plot.legend()
plot.show()
```

Как таковой закономерности тут не наблюдается. Просто доаольно резкие скачки от 1 до 4 эпизодов, именно в этом диапазоне кол-ва эпизодов находятся послдение 55 анимешек

```python
plot.plot(animes[animes['genre'] == 'Fantasy']['#'].head(55), animes[animes['genre'] == 'Fantasy']['rating'].head(55), label = 'рейтинг')
plot.title('Статистика рейтинга')
plot.ylabel('рейтинг')
plot.xlabel('ID аниме')
plot.legend()
plot.show()
```

Наглядно видно, что у первых 55 аниме с жанром Fantasy рейтинг стабильно идёт на спад. Можно проверить, связано ли это с кол-вом персонажей.

```python
plot.plot(animes[animes['genre'] == 'Fantasy']['#'].head(55), animes[animes['genre'] == 'Fantasy']['members'].head(55), label = 'кол-во персонажей')
plot.title('Статистика кол-ва персонажей')
plot.ylabel('кол-во ерсонажей')
plot.xlabel('ID аниме')
plot.legend()
plot.show()
```

В целом, трудно однозначно сказать, но видно, что количество персонажей у первых 55 аниме в жанре фентези стабильно падает, за исключением нескольких резких повышений значения это8го параметра. Быть может, дело было в этом. На верочку можно проверить и кличество эпизодов...

```python
plot.plot(animes[animes['genre'] == 'Fantasy']['#'].head(55), animes[animes['genre'] == 'Fantasy']['episodes'].head(55), label = 'кол-во эпизодов')
plot.title('Статистика кол-ва эпизодов')
plot.ylabel('кол-во эпизодов')
plot.xlabel('ID аниме')
plot.legend()
plot.show()
```

Есть некотрая связь: наблюдается волнообразный рост кол-ва персонажей на фоне прямого упадка рейтинга. Так что, эти явления никак не связаны...
